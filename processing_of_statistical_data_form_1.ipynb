```python
df_p = pd.read_csv('~/python/morbidity/Форма_1/population_form1_202408011803.csv', sep = ';')
```

```python
df.query("region == 'Брянская область' and nozol == 'Ветряная оспа' and year == 2022")
```

```python
df_all = df.merge(df_p, on = ['region', 'year', 'month'])
```

```python
df_all.head()
```

```python
df_all['ill_14_17_100k'] = (df_all.ill_17 - df_all.ill_14)*100000/df_all.population_14_17
```

```python
df_all['ill_over_17_100k'] = (df_all.ill_total - df_all.ill_17)*100000/df_all.population_over_17
```

```python
df_all.query("region == 'Брянская область' and nozol == 'Ветряная оспа' and year == 2022")
```

```python
df_all.to_excel('Вся информация.xlsx')
```

```python
df_final = df_all[['nozol', 'region', 'year', 'month', 'ill_total', 'ill_total_100k', 'ill_14', 'ill_14_100k', 'ill_14_17', 'ill_14_17_100k', 'ill_over_17', 'ill_over_17_100k']]
```
